# Restructuring Data

```python
weather = pd.read_csv('C:/Users/htc/Downloads/weather.csv')
```

```python
weather 
```

```python
# Many of the time we receive data in a way that we don't want it. So in these cases we have to restructure our data in a more beautiful way.
```

```python
# let's say this weather data has a wide format data. And I'm not comfortable with this type of data. How do I restructure it?
```

```python
weather.info()
# so we don't have null values here
```

```python
# Now let's restructure our data using pd.melt()
```

```python
# you see data is melted.
# let's change our variable name and value column name
pd.melt(weather,id_vars = ['year','month','element'])
```

```python
weather_melted = pd.melt(weather,id_vars = ['year','month','element'],var_name = 'days',value_name ='temperature')
```

```python
weather_melted
```

```python
# This is a temporary data if we like want our real data back, we can just get it from weather data set.
# However if you want to reverse it back using weather_melted we can do it.
# using pd.pivot()
```

```python
weather_pivoted = pd.pivot(weather_melted,index = ['year','month','element'],columns = 'days',values = 'temperature')
```

```python
weather_pivoted.reset_index()
```

```python
# Let's try to combine both melt and pivot in a single table. And see what we get.
```

```python
weather_tidy = pd.pivot(weather_melted,index = ['year','month','days'],
        columns = 'element', values = 'temperature')
```

```python
weather_tidy= weather_tidy.reset_index()
```

```python
weather_tidy.columns
```

```python
# let's remove this element, causing a confusion because of multi indexing
```

```python
weather_tidy.columns.name = None
```

```python
weather_tidy['days'] = weather_tidy['days'].apply(lambda x : x[3:])
```

```python
weather_tidy
```

```python
# now without using group by we want to calculate the mean of max and min elements based on days.
```

```python
# we will use pd.pivot_table
# By default it provides us aggfunc = mean
# if we have nan values we have one more parameter dropna = False
```

```python
pd.pivot_table(weather_melted, index=['year','month','days'],
              columns = 'element',values = ['temperature'] ).reset_index()
```

```python
df = pd.read_csv("C:/Users/htc/Downloads/Pfizer_1.csv")
```

```python
df.info()

# Here we can see some null values
```

```python
df_melted = pd.melt(df,id_vars=['Date','Drug_Name','Parameter'],
                   var_name = 'Time', value_name = 'Result')
```

```python
df_melted
```

```python
# Let's reverse it back
```

```python
df_pivoted = pd.pivot(df_melted,
                     index = ['Date','Drug_Name','Parameter'],
                     columns = 'Time',
                     values = 'Result')
```

```python
df_pivoted = df_pivoted.reset_index()
```

```python
df_pivoted.columns.name = None
```

```python
df_pivoted
```

```python
# Let's combine both pd.pivot and pd.melt
```

```python
df_tidy = pd.pivot(df_melted,
                  index=['Date','Drug_Name','Time'],
                  columns = 'Parameter',
                  values = 'Result').reset_index()
```

```python
df_tidy.columns.name=None
```

```python
df_tidy
```

```python
# Let's say if we want to calculate the mean of pressure and temperature for every day and for every drug. Without using Group by.
```

```python
pd.pivot_table(df_tidy,
              index = ['Drug_Name'],
              columns = 'Date',
              values = ['Temperature'],
              aggfunc = 'mean')
```

```python
pd.pivot_table(df_tidy,
              index = ['Drug_Name'],
              columns = 'Date',
              values = ['Temperature'],
              aggfunc = 'mean',margins = True)
```
