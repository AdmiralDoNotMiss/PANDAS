# Grouping

```python
movies = pd.read_csv('C:/Users/htc/movies.csv')
```

```python
import pandas as pd
```

```python
directors = pd.read_csv('C:/Users/htc/directors.csv')
```

```python
movies.drop(columns = 'Unnamed: 0',inplace  = True)

directors.drop(columns = 'Unnamed: 0',inplace  = True)
```

```python
data = movies.merge(directors,left_on='director_id',right_on='id')
```

```python
data.drop(columns=['id_y','director_id'],inplace=True)
```

```python
data['budget'] = (data['budget']/1000000).round(2)
```

```python
data['revenue'] = (data['revenue']/1000000).round(2)
```

```python
# If I wish to see how many groups are there?

data.groupby('director_name').ngroups
```

```python
# If you wish to see all the groups?

data.groupby('director_name').groups

# This returns a dictionary, where director_names are key and the index(where director_name) is present are values
```

```python
# Get the groups in a dataframe for a particular director.

data.groupby('director_name').get_group('Kevin Reynolds')

# Confirmed that .groups attribute gave us accurate indices
# Check for Kevin Reynolds in above row
```

```python
# Count How many movies did each director make?
# without using value_counts()

data.groupby('director_name')['title'].count().to_frame()
```

```python
# Get the very first and last movie created by each director

data.groupby('director_name')['year'].agg(['min','max'])
```

```python
# Important topic
# Get me all the high budget directors (budget>100 mill). There is one condition, fetch all the movies of that particular director.
# DO IT MANUALLY
```

```python
dir_budgets = data.groupby('director_name')['budget'].agg(['max']).reset_index()
```

```python
high_budget_dirs = dir_budgets.loc[dir_budgets['max']>=100]['director_name']
```

```python
high_budget_dirs
```

```python
# Now how do we fetch all the movies
# We will use isin method

data.loc[data['director_name'].isin(high_budget_dirs)]
```

```python
# isn't this a very tidious method?
# Can we do it in one line
# just like we used to do it in sql group by + having
```

```python
data.groupby('director_name').filter(lambda x: x['budget'].max()>=100)

# same results
```

```python
# create a new column for every director and tell, if any particular movie is risky or not?
# Risky movie means(budget of the movie > avg revenue of the director)
```

```python
# first we are creating a custom function

def my_func(x):
    x['risky'] = x['budget'] - x['revenue'].mean() >=0
    return x

risky_movies = data.groupby('director_name').apply(my_func)

# This will tell us if a movie is risky or not.
# But we want movies that were risky
```

```python
risky_movies.loc[risky_movies['risky']]
```

```python
# How do we select multiple columns for Aggregation

data_multi = data.groupby('director_name')[['year','title']].agg({'year':['min','max'],
                                                  'title':['count']})

# This concept is introducing Multi level indexing.
# Year is level 1 index'
# min,max & count is level 2 indexing
```

```python
data_multi
```

```python
data_multi['year']
```

```python
# Is there a way to fix this?

data_multi.columns = ["_".join(col) for col in data_multi.columns]
```

```python
data_multi.reset_index()
```

```python
# There is one more way to do it directly. Don't have to do it in such a hassle

data_multi = data.groupby('director_name').agg(year_min=('year','min'),
                                 year_max = ('year','max'),
                                 title_count = ('title','count')).reset_index()
```

```python
data_multi
```

```python
# Who are the most productive directors?

data_multi['years_active'] = data_multi['year_max'] - data_multi['year_min']

```

```python
data_multi['movies_per_year'] = data_multi['title_count']/data_multi['years_active']
```

```python
data_multi.sort_values(by = ['movies_per_year'],ascending = False)
```

```python
# You have a data frame that consists of the names and ages of a few candidates. You need to convert age into categories according to the following conditions:
# 1. If age<18: kid
# 2. age >=18 and <=50: adult
# 3. age>50 senior
# Write a python program to do so using apply function.
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'name':["Ram","Shyam","Mukesh","Suresh"],'age':[10,18,60,50]})
df
```

```python
def categorize_age(age):
    if age < 18:
        return 'kid'
    elif age <= 50:
        return 'adult'
    else:
        return 'senior'

```

```python
df['age'] = df['age'].apply(categorize_age)
```
