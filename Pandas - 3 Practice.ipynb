# Data Cleaning


```python
df.duplicated().tail(15)

# Just like masking it returns us the boolean values. True for duplicates and false for unique.
# PS ~ First duplicate will be a unique value(False)
```

```python
df.loc[df.duplicated()].tail(15)
```

```python
# Now let's remove duplicates and cleanse our  data
# df.drop_duplicates function. in parameter we have keep. we can pass 3 commands to keep
# Keep = 'first' this will keep the first occurrence only
# Keep = 'last' this will keep the last occurrence only
# Keep = 'False' this will delete all the duplicates including the first value 

df.drop_duplicates(keep='first')
```

```python
df.drop_duplicates(keep='last')
```

```python
df.drop_duplicates(keep=False)
```
